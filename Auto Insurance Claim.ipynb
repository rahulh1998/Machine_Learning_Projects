```python
#importing essential libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns',70)
```

## Data Preparation

```python
# Checking class imbalance
df['target'].value_counts()
```

```python
#dropping column 'id'  
df.drop('id',1,inplace = True)
```

```python
# checking for null values %

(df.isnull().sum()/df.shape[0])*100
```

```python
# dropping all the columns having null value% greater than 18%

for x in df.columns:
    if (df[x].isnull().sum()/df.shape[0])*100 >18:
        df.drop(x,1,inplace=True)
```

```python
# Listing missing values% > 0

miss_cols = []
for x in df.columns:
    if (df[x].isnull().sum()/df.shape[0])*100 > 0:
        miss_cols.append(x)
        print(f'{x:20}  {(df[x].isnull().sum()/df.shape[0])*100:>20}')
```

```python
df[miss_cols].head()
```

```python
# Missing values imputation 

miss_cat = miss_cols[:-3]
miss_num = miss_cols[6:]

from sklearn.impute import SimpleImputer

#for categorical variables
si_c = SimpleImputer(missing_values=np.nan , strategy='most_frequent')
df[miss_cat] = si_c.fit_transform(df[miss_cat])

#for numerical variables
si_n = SimpleImputer(missing_values=np.nan , strategy='mean')
df[miss_num] = si_n.fit_transform(df[miss_num])
```

```python
#df.isnull().sum() 
```

```python
# Data type conversion for categorical variables

cat_cols = []
for i in df.columns:
    if 'cat' in str(i):
        cat_cols.append(i)
        
cat_cols
```

```python
for i in cat_cols:
    df[i] = df[i].astype('int64')
```

```python
df[cat_cols].info()
```

```python
# Creating X & y 
X = df.drop('target',1)
y = df['target']
```

```python
# Spillting the dataset into train & test

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.3 , random_state=101,stratify=y)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Resampling using smote 
from imblearn.over_sampling import SMOTE
sm = SMOTE()
X_train , y_train = sm.fit_resample(X_train,y_train)
```

```python
# Scaling the features

cols = X_train.select_dtypes(exclude='int64').columns

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train[cols] = ss.fit_transform(X_train[cols])
X_train.head()
```

# Model Building

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)

y_pred_rf = rf.predict(X_test)

from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred_rf))
```
